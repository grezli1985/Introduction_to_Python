3. Найти интервалы, на которых ф-ция убывает: (-9.9789538; -6.83137) u (-4.167784; -0.39267) u (0.45457; 1.70061 ) u (3.81931; 7.0010317) u (9.877149; +oo)

```python
import sympy
from sympy.plotting import plot
```

```python
x = sympy.Symbol('x')
print(x)
```

```python
f_x = -12 * x**4 * sympy.sin(sympy.cos(x)) - 18 * x**3 + 5 * x**2 + 10 * x - 30
print(f_x)
```

```python
print (f_x.subs(x, 10.0))
```

```python
# plot(f_x) #, ylim=(-100, 100), xlim=(-10, 10))
plot(f_x, ylim=(-100, 100), xlim=(-10, 10))
```
