Дана функция f(x) = 5x^2 + 10x - 30

   1. Определить корни (1.64575131106459,  -3.64575131106459)

   2. Найти интервалы, на которых функция возрастает (-1, oo)

   3. Найти интервалы, на которых функция убывает (-oo, -1)

  4. Построить график

    5. Вычислить вершину (-1, 35)

   6. Определить промежутки, на котором f > 0 (x < -3.64575131106459 && x > 1.64575131106459)

    7. Определить промежутки, на котором f < 0

```python
import sympy
from sympy.plotting import plot
```

```python
x = sympy.Symbol('x')
print(x)
```

```python
f_x = 5 * x**2 + 10 * x - 30
print(f_x)
```

```python
for root in sympy.solve(f_x, x):
    print(root.evalf())
```

```python
f_diff_x = f_x.diff()
print(sympy.solve(f_diff_x))
```

```python
print(f_x.subs(x, -1))
```

```python
plot(f_x, f_diff_x) #, xlim=(0, 1))
```

```python
plot(f_diff_x) #, xlim=(0, 1))
```

```python
for i in range(560, 580):
    param = i / 1000
    print(param, f_diff_x.subs(x, param))
```

```python
print(sympy.solve(f_diff_x, x)[0].evalf())
```
